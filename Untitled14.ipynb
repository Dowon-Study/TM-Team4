```python
import os
import sys
import urllib.request
import datetime
import time
import json

client_id = "qBNipzbswul0f2qrOlEm"
client_secret = "_N8nvjsxbQ"

# [CODE 1]
# url 받고 요청하는 함수
def getRequestUrl(url):
  req = urllib.request.Request(url)
  req.add_header("X-Naver-Client-Id", client_id)
  req.add_header("X-Naver-Client-Secret", client_secret)

  try:
    response = urllib.request.urlopen(req)
    if response.getcode() == 200:
      print("[%s] Url Request Success" % datetime.datetime.now())
      return response.read().decode('utf-8')
  except Exception as e:
    print(e)
    print("[%s] Error for URL : %s" % (datetime.datetime.now(), url))
    return None

# [CODE 2]
def getNaverSearch(node, srcText, start, display, start_date, end_date):
  base = "https://openapi.naver.com/v1/search"  # 코드1에 보낼 url 제작
  node = "/%s.json" % node
  parameters = "?query=%s&start=%s&display=%s" % (urllib.parse.quote(srcText), start, display)

  url = base + node + parameters  # url에 마지막 파라미터를 받고 url 주소 완성
  responseDecode = getRequestUrl(url)  # 완성된 주소를 코드1에 보내고 그 값을 변수에 저장

  if (responseDecode == None):
    return None
  else:
    return json.loads(responseDecode)

# [CODE 3]
# 받고 싶은 정보를 변수값에 저장
def getPostData(post, jsonResult, cnt):
  title = post['title'] # 뉴스 제목
  description = post['description'] # 메인 내용
  org_link = post['originallink'] # 뉴스 페이지 원본 링크
  link = post['link'] # 뉴스 페이지 링크

  # 시간 데이터 및 위에 끌어온 정보들을 최종 결과값인 jsonResult에 append 추가
  pDate = datetime.datetime.strptime(post['pubDate'], '%a, %d %b %Y %H:%M:%S +0900')
  pDate = pDate.strftime('%Y-%m-%d %H:%M:%S')

  jsonResult.append({'cnt': cnt, 'title': title, 'description': description, 'org_link': org_link, 'link': org_link, 'pDate': pDate})
  return

def generate_date_ranges():  # 검색 기간을 하루 단위로 분할하는 함수
    start_date = datetime.datetime(2025, 1, 1)
    end_date = datetime.datetime.today()
    date_ranges = []
    while start_date < end_date:
        next_day = start_date + datetime.timedelta(days=1)
        date_ranges.append((start_date.strftime('%Y%m%d'), next_day.strftime('%Y%m%d')))
        start_date = next_day
    return date_ranges

# [CODE 4]
def main():
  node = 'news' # 무엇을 크롤링할 지 결정하는 변수
  srcText = input('검색어를 입력하세요: ')
  cnt = 0
  jsonResult = [] # 위 모든 데이터들을 받는 리스트
  date_ranges = generate_date_ranges()

  for start_date, end_date in date_ranges:
    jsonResponse = getNaverSearch(node, srcText, 1, 100, start_date, end_date)
    if jsonResponse is None:
      continue
    total = jsonResponse.get('total', 0)

    # 데이터를 하나씩 계속 받는 반복문
    while ((jsonResponse != None) and (jsonResponse['display'] != 0)): # 크롤링할때 데이터가 없거나 검색 결과가 없는 경우 크롤링 실행x
      for post in jsonResponse['items']:
        cnt += 1
        getPostData(post, jsonResult, cnt) # [CODE 3]

      start = jsonResponse['start'] + jsonResponse['display']
      if start > 1000: # 네이버 API는 1000 까지만 허용
        break
      jsonResponse = getNaverSearch(node, srcText, start, 100, start_date, end_date) #[CODE 2]

  print('전체 검색 : %d 건' % total)

  with open('%s_naver_%s.json' % (srcText, node), 'w', encoding='utf8') as outfile: # with open 함수를 통해 파일 작성 / 문자 깨지는 오류 방지
    jsonFile = json.dumps(jsonResult, indent=4, sort_keys=True, ensure_ascii=False)

    outfile.write(jsonFile)

  print("가져온 데이터 : %d 건" % (cnt))
  print('%s_naver_%s.json SAVED' % (srcText, node))

if __name__ == '__main__':
  main() # 파일을 실행시키는 곳이 이 파일이랑 동일시할 때 실행
```
